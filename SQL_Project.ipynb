```python
##First 5 rows of the billionaires dataset
%%sql 
SELECT * FROM billionaires 
LIMIT 5
```

```python
##Youngest Billionaire

%sql 
SELECT * FROM billionaires 
WHERE age IN (SELECT MIN(age) FROM billionaires)
```

```python
##Oldest Billionaire

%sql 
SELECT * FROM billionaires 
WHERE age IN (SELECT MAX(age) FROM billionaires)
```

```python
## Number of billionaires that made it to the 2022 list
%sql 
SELECT COUNT(name) 
FROM billionaires
```

```python
## Number of billionaires in each country
%sql SELECT country, 
COUNT(country) as Number_of_Billionaires 
FROM billionaires 
GROUP BY country 
ORDER BY COUNT(country) desc
```

```python
## How many billionaires are self made

%sql 
SELECT self_made, 
COUNT(self_made) 
FROM billionaires 
GROUP BY self_made
```

```python
## Number of billionaires according to Gender

%sql 
SELECT gender, 
COUNT(gender) 
FROM billionaires 
WHERE gender <> 'None' 
GROUP BY gender
```

```python
##Billionaires in Nigeria

%sql 
SELECT * 
FROM billionaires 
WHERE country = 'Nigeria'
```

```python
##Top billionaires in each industry category

%sql 
SELECT * 
FROM (SELECT name, category,final_worth_billion_dollars, 
      MAX(final_worth_billion_dollars) OVER (PARTITION BY category) 
      AS Max_Billions FROM billionaires)A 
WHERE final_worth_billion_dollars = Max_Billions 
ORDER BY final_worth_billion_dollars DESC
```

```python
##Number of billionaires in each industry category

%sql 
SELECT category, 
COUNT(category) 
FROM billionaires 
GROUP BY category 
ORDER BY COUNT(category) DESC
```

```python
##common titles billionaires carry
%sql 
SELECT title, 
COUNT(title) 
FROM billionaires 
GROUP BY title 
ORDER BY COUNT(title) DESC
```
